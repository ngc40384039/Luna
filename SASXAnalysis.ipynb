```python
import numpy as np
import matplotlib.pyplot as plt
#LOADING DATA
data = np.loadtxt("/content/PEG1_1.dat")  #columns = [q, I(q)]
q = data[:, 0]  #Scattering vector 1/nm
Iq = data[:, 1]  #Intensity a.u.
#PLOTTING q VS I
plt.figure(figsize=(8, 6))
plt.loglog(q, Iq, marker='o', linestyle='-', color='blue', label='SAXS Data')
plt.title("SAXS Profile: I(q) vs q")
plt.xlabel("q (1/nm)")
plt.ylabel("I(q) (a.u.)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
